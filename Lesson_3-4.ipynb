```python
t = np.linspace(0,10,51) 
f = np.cos(t)
print(t)
print(f)
plt.plot(t,f, c = "green")
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5 , 2.5])
plt.show()
```

```python
x = np.linspace(-3,3,51) 
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График 1')
ax2.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График 2')

ax3.plot(x, y3)
ax3.set_title('График 3')

ax4.plot(x, y4)
ax4.set_title('График 4')

plt.show()
```

```python
credit_cards = pd.read_csv("creditcard.csv", sep=",")

credit_cards.head()
```

```python
plt.style.use('fivethirtyeight')

class_list = credit_cards['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind="bar")
plt.show()
```

```python
class_list.plot(kind="bar", logy = True)
plt.show()
```

```python
v1_class0 = credit_cards.loc[credit_cards['Class'] == 0, ['V1']]
v1_class1 = credit_cards.loc[credit_cards['Class'] == 1, ['V1']]

plt.hist(v1_class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color ='grey')
plt.hist(v1_class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color ='red')
plt.legend()
plt.show()
```
